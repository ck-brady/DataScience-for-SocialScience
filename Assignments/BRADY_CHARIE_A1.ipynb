```python
# example of f-string use
firstname="fabien"
length_first_name=len(firstname)
print(f"There are {length_first_name} letters in {firstname.capitalize()}.")
```

```python
#1. Create a dictionary
mydict={}
grades = ["F","E","D","D+","C","C+","B","B+","A-","A","A+"]

for key, value in enumerate(grades):
    mydict[key] = value

mydict
```

```python
#2. define the function

def letter_grade(number):
    if number in [0,1]:
        comment = "Sorry!"
    elif number in [2,3]:
        comment = "You passed!"
    elif number in [4,5]:
        comment = "Good!"
    elif number in [6,7]:
        comment = "Congratulations!"
    elif number in [8,9,10]:
        comment = "Amazing!"
    print(f"{comment} Your grade {number} means you received {mydict[number]} for this course")
```
