```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
import sklearn.linear_model as skl_lm

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
titanicPath="C:/Users/16136/Desktop/Fall 2021/ECO4199"
```

```python
df=pd.read_csv('/'.join([titanicPath,'titanic.csv']))
df['nameLength'] = df['Name'].str.len()
df.head()
```

### Model 1

```python
mod1 = smf.ols('Fare ~ nameLength', data=df).fit()
print(mod1.summary())
```

Interpretation:

#### 1. Interpret the coefficient values

The intercept $\hat{\beta}_0 = 7.7054$ is not especially meaningful. It is the value when name length is zero, but nobody has a name of zero length. A realistic minimum name length would possibly be around 11-12 (4-5 digits for title, 3 letter first name, 1 digit space, 3 letter last name), so the intercept cannot be interpreted as the fare in the absence of name length.

Since $\hat{\beta}_1 = 0.9715$, a one digit increase in name length increases the fare price by ~97 cents. This is a positive relationship between name length and fare price, such that people with longer names typically paid larger fares.

#### 2. Interpret their significance

Given that $\rho = 0 < \alpha = 0.05$, the association between name length and fare is statistically significant at the 95% confidence level. 

####  $\hat{\beta}_1 = \frac{cov(Y,X)}{var(X)}$

```python
Syx = df[['Fare','nameLength']].cov().iloc[0,1] # covariance
Vx = df['nameLength'].var()
beta_1 = Syx/Vx
print(f"The value for the slope obtained using OLS: {mod1.params['nameLength']} is the same as the one obtained from computing the covariance of X and Y over the variance of X: {beta_1}" )
      
```

### Model 2

```python
mod2 = smf.ols('Fare ~ Age', data=df).fit()
print(mod2.summary())
```

Interpretation:


#### 1. Interpret the coefficient values

The intercept $\hat{\beta}_0 = 20.6354$ is the predicted fare price when the passenger's age is zero. Again, this is not especially meaningful in this case given that infants do not typically have their own fare rates that are different than a child, teen, etc. 

Since $\hat{\beta}_1 = 0.396$, a one year increase in age is associated with an increase in the fare price by ~40 cents. This is a positive relationship between age and fare price, such that older passengers typically paid larger fares than younger passengers.

#### 2. Interpret their significance

Given that $\rho = 0.001 < \alpha = 0.05$, the association between age and fare is statistically significant at the 95% confidence level. This does not mean that the effect is large, only that it is statistically different from zero.

### Model 4

```python
df['Pclass'] = df['Pclass'].astype("category")

mod4 = smf.ols('Fare ~ Pclass', data=df).fit()
print(mod4.summary())
```

Interpretation:

#### 1. Interpret the coefficient values

The intercept $\hat{\beta}_0 = 84.1547$ is the predicted fare price for a passenger in first class.

The coefficient $\hat{\beta}_1 = -63.4925$ is the difference in fare price between a second class passenger and a first class passenger. It is negative, meaning second class passengers have lower fares. A second class passenger's predicted fare price is $\hat{\beta}_0 + \hat{\beta}_1$. 

The coefficient $\hat{\beta}_2 = -70.447$ is the difference in fare price between a third class passenger and a first class passenger. It is negative, meaning third class passengers have lower fares, which is also lower than second class passengers. A third class passenger's predicted fare price is $\hat{\beta}_0 + \hat{\beta}_2$. 

#### 2. Interpret their significance

Given that $\rho_1 = \rho_2 = 0 < \alpha = 0.05$, fare prices of second and third class passengers are significantly different than first class passengers at the 95% confidence level. This conclusion is also supported by the low p-value of the F-statistic, which tests whether the coefficients are jointly zero.

- Using the pacakge [Statsmodels](https://www.statsmodels.org/stable/example_formulas.html) and OLS regression (linear probability model), regress `Survived` on `Pclass  + Sex + Age + Fare + nameLength`


- Repeat the operation using the [Probit model](https://www.statsmodels.org/stable/generated/statsmodels.formula.api.probit.html)


- Note that you will need to create dummy variables for `Pclass` and `Sex` using Pandas [get_dummies](https://pandas.pydata.org/docs/reference/api/pandas.get_dummies.html) function. You only need to do this once for both models
   
   
- Compare, in a markdown cell, the mean absolute errors of both models

```python
# Create dummies

dummies = pd.get_dummies(df[['Pclass', 'Sex']], drop_first=True)
df2 = df.join(dummies)
df2.head()
```

### Linear Probability Model

```python
mod10 = smf.ols('Survived ~ Pclass_2 + Pclass_3 + Sex_male + Age + Fare + nameLength', data=df2).fit() 
print(mod10.summary())
```

```python
Y_pred = mod10.predict(df2[['Pclass_2','Pclass_3','Sex_male','Age','Fare','nameLength']])
Y_true = df2['Survived'].values
mae = mean_absolute_error(Y_true, Y_pred)
print(f"The in-sample MAE of this model is: {mae}")
```

### Logit (as per email)

```python
mod11 = smf.logit('Survived ~ Pclass_2 + Pclass_3 + Sex_male + Age + Fare + nameLength', data=df2).fit()
print(mod11.summary())
```

```python
Y_pred = mod11.predict(df2[['Pclass_2','Pclass_3','Sex_male','Age','Fare','nameLength']])
Y_true = df2['Survived'].values
mae = mean_absolute_error(Y_true, Y_pred)
print(f"The in-sample MAE of this model is: {mae}")
```

### Compare the mean absolute errors of both models

The MAE for the logistic regression model is smaller than the MAE for the linear probability model (0.288 vs. 0.452).  The logit model is a much better predictor for Survived.
